```python
# Convert Date formate
df['Submission_date'] = pd.to_datetime(
    df['Submission_date'],
    errors='coerce'
)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('default')
```

##**Candidate Status Distribution**

```python
df['Status'].value_counts()
```

```python
df['Status'].value_counts().plot(
    kind='bar',
    figsize=(8,5)
)

plt.title('Candidate Status Distribution')
plt.ylabel('Count')
plt.show()
```

##**Top Recruitment Sources**

```python
top_sources = df['Source'].value_counts()

print(top_sources)
```

```python
top_sources.plot(
    kind='bar',
    figsize=(8,5)
)

plt.title('Candidates by Source')
plt.ylabel('Count')
plt.show()
```

##**Client Wise Submissions**

```python
client_submission = df.groupby('Client').size()

print(client_submission.sort_values(ascending=False))
```

```python
client_submission.sort_values(ascending=False).head(10).plot(
    kind='bar',
    figsize=(10,5)
)

plt.title('Top Clients by Submission')
plt.ylabel('Count')
plt.show()
```

##**Salary Distribution**

```python
plt.figure(figsize=(8,5))

sns.histplot(
    df['Salary'],
    bins=20,
    kde=True
)

plt.title('Salary Distribution')
plt.show()
```

##**Salary Outlier Detection**

```python
plt.figure(figsize=(8,4))

sns.boxplot(x=df['Salary'])

plt.title('Salary Outliers')
plt.show()
```

##**Interview Trend Over Time**

```python
df['Interview_date'] = pd.to_datetime(df['Interview_date'])

interview_trend = (
    df.groupby(df['Interview_date'].dt.month)
      .size()
)

print(interview_trend)
```

```python
interview_trend.plot(
    kind='line',
    marker='o',
    figsize=(8,5)
)

plt.title('Monthly Interviews')
plt.xlabel('Month')
plt.ylabel('Count')
plt.show()
```
